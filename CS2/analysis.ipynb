```python
import pandas as pd
import numpy as np
import warnings
from collections import defaultdict

warnings.filterwarnings("ignore")


all_games = pd.read_csv("cs2_all_tiers_games.csv")
all_games["datetime"] = pd.to_datetime(all_games["datetime"], errors="coerce")

all_games = all_games.sort_values("datetime").reset_index(drop=True)



maps_raw = all_games[
    (all_games["is_total"] == False) &
    (all_games["map_name"].notna()) &
    (all_games["datetime"].notna())
].copy()

maps_raw = maps_raw.sort_values("datetime").reset_index(drop=True)
maps_raw["team1_win"] = maps_raw["team1_win"].astype(int)

print("maps_raw shape:", maps_raw.shape)
print("\nOriginal map target distribution:")
print(maps_raw["team1_win"].value_counts(normalize=True))
print(maps_raw["team1_win"].value_counts())



def compute_players_avg(games_data, player_ids):
    games_data = games_data.copy()
    games_data = games_data[
        (games_data["is_total"] == False) &
        (games_data["datetime"].notna())
    ]

    all_players = []

    for i in range(1, 3):
        for j in range(1, 6):
            prefix = f"team{i}_player{j}"

            needed_cols = [
                f"{prefix}_id",
                prefix,
                f"{prefix}_kast",
                f"{prefix}_kills",
                f"{prefix}_deaths",
                f"{prefix}_adr",
                f"{prefix}_kddiff"
            ]

            existing_cols = [c for c in needed_cols if c in games_data.columns]

            data = games_data[existing_cols].copy()

            rename_map = {
                f"{prefix}_id": "player_id",
                prefix: "player",
                f"{prefix}_kast": "kast",
                f"{prefix}_kills": "kills",
                f"{prefix}_deaths": "deaths",
                f"{prefix}_adr": "adr",
                f"{prefix}_kddiff": "kddiff"
            }

            data = data.rename(columns=rename_map)

            for col in ["player_id", "player", "kast", "kills", "deaths", "adr", "kddiff"]:
                if col not in data.columns:
                    data[col] = np.nan

            all_players.append(data)

    all_players = pd.concat(all_players, ignore_index=True)
    all_players = all_players.dropna(subset=["player_id"])

    all_players["player_id"] = all_players["player_id"].astype(float)
    player_ids = [float(x) for x in player_ids if pd.notna(x)]

    all_players = all_players[all_players["player_id"].isin(player_ids)]

    if len(all_players) == 0:
        return pd.DataFrame(columns=[
            "player_id",
            "player",
            "avg_kast",
            "avg_adr",
            "avg_kddiff",
            "total_kills",
            "total_deaths",
            "maps_played",
            "kd_ratio",
            "player_order"
        ])

    final_df = (
        all_players
        .groupby("player_id", as_index=False, sort=False)
        .agg(
            player=("player", "first"),
            avg_kast=("kast", "mean"),
            avg_adr=("adr", "mean"),
            avg_kddiff=("kddiff", "mean"),
            total_kills=("kills", "sum"),
            total_deaths=("deaths", "sum"),
            maps_played=("player", "count")
        )
    )

    final_df["kd_ratio"] = (
        final_df["total_kills"] /
        final_df["total_deaths"].replace(0, np.nan)
    )

    order = {float(pid): i for i, pid in enumerate(player_ids)}
    final_df["player_order"] = final_df["player_id"].map(order)

    final_df = final_df.sort_values("player_order")

    return final_df


def compute_team_avgs_before_time(time, player_ids):
    current_games_df = all_games[
        all_games["datetime"] < pd.Timestamp(time)
    ]

    return compute_players_avg(current_games_df, player_ids)



player_map_index = defaultdict(lambda: defaultdict(list))

for _, row in maps_raw.iterrows():
    dt = row["datetime"]
    map_name = row["map_name"]

    for team_num in [1, 2]:
        if team_num == 1:
            won = int(row["team1_win"])
        else:
            won = int(1 - row["team1_win"])

        for player_num in range(1, 6):
            prefix = f"team{team_num}_player{player_num}"
            pid = row.get(f"{prefix}_id")

            if pd.notna(pid):
                player_map_index[float(pid)][map_name].append({
                    "datetime": dt,
                    "won": won
                })


def get_team_map_winrate_by_players(player_ids, map_name, time, days_back=365):
    cutoff = pd.Timestamp(time)
    since = cutoff - pd.Timedelta(days=days_back)

    winrates = []

    for pid in [float(p) for p in player_ids if pd.notna(p)]:
        games = [
            g for g in player_map_index[pid].get(map_name, [])
            if since <= g["datetime"] < cutoff
        ]

        if games:
            winrates.append(np.mean([g["won"] for g in games]))

    return np.mean(winrates) if winrates else np.nan



player_history = defaultdict(list)

for _, row in maps_raw.iterrows():
    dt = row["datetime"]
    map_name = row["map_name"]

    for team_num in [1, 2]:
        if team_num == 1:
            team_id = row["team1_id"]
            won = int(row["team1_win"])
        else:
            team_id = row["team2_id"]
            won = int(1 - row["team1_win"])

        for player_num in range(1, 6):
            prefix = f"team{team_num}_player{player_num}"

            pid = row.get(f"{prefix}_id")

            if pd.isna(pid):
                continue

            kills = row.get(f"{prefix}_kills", np.nan)
            deaths = row.get(f"{prefix}_deaths", np.nan)
            adr = row.get(f"{prefix}_adr", np.nan)
            kast = row.get(f"{prefix}_kast", np.nan)
            kddiff = row.get(f"{prefix}_kddiff", np.nan)

            kd = (
                kills / deaths
                if pd.notna(kills) and pd.notna(deaths) and deaths != 0
                else np.nan
            )

            player_history[float(pid)].append({
                "datetime": dt,
                "team_id": team_id,
                "map_name": map_name,
                "won": won,
                "kills": kills,
                "deaths": deaths,
                "adr": adr,
                "kast": kast,
                "kd": kd,
                "kddiff": kddiff
            })


def empty_player_form():
    return {
        "adr": np.nan,
        "kast": np.nan,
        "kd": np.nan,
        "kddiff": np.nan,
        "wr": np.nan,
        "maps": 0,
        "form_score": np.nan
    }


def get_player_recent_form(player_id, time, last_n=10, map_name=None):
    if pd.isna(player_id):
        return empty_player_form()

    cutoff = pd.Timestamp(time)

    history = player_history.get(float(player_id), [])

    games = [
        g for g in history
        if g["datetime"] < cutoff and (map_name is None or g["map_name"] == map_name)
    ]

    if len(games) == 0:
        return empty_player_form()

    games = sorted(games, key=lambda x: x["datetime"])[-last_n:]

    adr = np.nanmean([g["adr"] for g in games])
    kast = np.nanmean([g["kast"] for g in games])
    kd = np.nanmean([g["kd"] for g in games])
    kddiff = np.nanmean([g["kddiff"] for g in games])
    wr = np.nanmean([g["won"] for g in games])
    maps = len(games)

    form_score = (
        0.40 * kd +
        0.30 * (adr / 80) +
        0.20 * (kast / 75) +
        0.10 * wr
    )

    return {
        "adr": adr,
        "kast": kast,
        "kd": kd,
        "kddiff": kddiff,
        "wr": wr,
        "maps": maps,
        "form_score": form_score
    }


def get_team_player_form(player_ids, time, last_n=10, map_name=None):
    forms = [
        get_player_recent_form(pid, time, last_n=last_n, map_name=map_name)
        for pid in player_ids
        if pd.notna(pid)
    ]

    if len(forms) == 0:
        return empty_player_form()

    return {
        "adr": np.nanmean([f["adr"] for f in forms]),
        "kast": np.nanmean([f["kast"] for f in forms]),
        "kd": np.nanmean([f["kd"] for f in forms]),
        "kddiff": np.nanmean([f["kddiff"] for f in forms]),
        "wr": np.nanmean([f["wr"] for f in forms]),
        "maps": np.nanmean([f["maps"] for f in forms]),
        "form_score": np.nanmean([f["form_score"] for f in forms])
    }



records = []

for _, game in maps_raw.iterrows():

    match_id = game["match_id"]
    game_id = game["game_id"]
    game_time = game["datetime"]
    map_name = game["map_name"]

    team1_player_ids = [
        game[f"team1_player{j}_id"]
        for j in range(1, 6)
        if pd.notna(game.get(f"team1_player{j}_id"))
    ]

    team2_player_ids = [
        game[f"team2_player{j}_id"]
        for j in range(1, 6)
        if pd.notna(game.get(f"team2_player{j}_id"))
    ]

    team1_stats = compute_team_avgs_before_time(game_time, team1_player_ids)
    team2_stats = compute_team_avgs_before_time(game_time, team2_player_ids)


    t1_kast = team1_stats["avg_kast"].mean()
    t1_kd = team1_stats["kd_ratio"].mean()
    t1_adr = team1_stats["avg_adr"].mean()
    t1_kddiff = team1_stats["avg_kddiff"].mean()
    t1_maps = team1_stats["maps_played"].mean()

    t1_kills_map = (
        team1_stats["total_kills"] /
        team1_stats["maps_played"].replace(0, np.nan)
    ).mean()

    t1_deaths_map = (
        team1_stats["total_deaths"] /
        team1_stats["maps_played"].replace(0, np.nan)
    ).mean()

    t2_kast = team2_stats["avg_kast"].mean()
    t2_kd = team2_stats["kd_ratio"].mean()
    t2_adr = team2_stats["avg_adr"].mean()
    t2_kddiff = team2_stats["avg_kddiff"].mean()
    t2_maps = team2_stats["maps_played"].mean()

    t2_kills_map = (
        team2_stats["total_kills"] /
        team2_stats["maps_played"].replace(0, np.nan)
    ).mean()

    t2_deaths_map = (
        team2_stats["total_deaths"] /
        team2_stats["maps_played"].replace(0, np.nan)
    ).mean()


    t1_map_wr = get_team_map_winrate_by_players(
        team1_player_ids,
        map_name,
        game_time,
        days_back=365
    )

    t2_map_wr = get_team_map_winrate_by_players(
        team2_player_ids,
        map_name,
        game_time,
        days_back=365
    )

    row = {
        "match_id": match_id,
        "game_id": game_id,
        "datetime": game_time,
        "map_name": map_name,

        "team1_map_win": int(game["team1_win"]),

        "team1_avg_kast": t1_kast,
        "team1_avg_kd": t1_kd,
        "team1_avg_adr": t1_adr,
        "team1_avg_kddiff": t1_kddiff,
        "team1_avg_maps_played": t1_maps,
        "team1_kills_per_map": t1_kills_map,
        "team1_deaths_per_map": t1_deaths_map,

        "team2_avg_kast": t2_kast,
        "team2_avg_kd": t2_kd,
        "team2_avg_adr": t2_adr,
        "team2_avg_kddiff": t2_kddiff,
        "team2_avg_maps_played": t2_maps,
        "team2_kills_per_map": t2_kills_map,
        "team2_deaths_per_map": t2_deaths_map,

        "diff_avg_kast": t1_kast - t2_kast,
        "diff_avg_kd": t1_kd - t2_kd,
        "diff_avg_adr": t1_adr - t2_adr,
        "diff_avg_kddiff": t1_kddiff - t2_kddiff,
        "diff_avg_maps_played": t1_maps - t2_maps,
        "diff_kills_per_map": t1_kills_map - t2_kills_map,
        "diff_deaths_per_map": t1_deaths_map - t2_deaths_map,

        "team1_map_winrate": t1_map_wr,
        "team2_map_winrate": t2_map_wr,
        "diff_map_winrate": (
            (t1_map_wr if not np.isnan(t1_map_wr) else 0.5) -
            (t2_map_wr if not np.isnan(t2_map_wr) else 0.5)
        )
    }


    for last_n in [5, 10, 20]:
        t1_form = get_team_player_form(
            team1_player_ids,
            game_time,
            last_n=last_n,
            map_name=None
        )

        t2_form = get_team_player_form(
            team2_player_ids,
            game_time,
            last_n=last_n,
            map_name=None
        )

        row[f"team1_player_adr_form_{last_n}"] = t1_form["adr"]
        row[f"team2_player_adr_form_{last_n}"] = t2_form["adr"]
        row[f"diff_player_adr_form_{last_n}"] = t1_form["adr"] - t2_form["adr"]

        row[f"team1_player_kast_form_{last_n}"] = t1_form["kast"]
        row[f"team2_player_kast_form_{last_n}"] = t2_form["kast"]
        row[f"diff_player_kast_form_{last_n}"] = t1_form["kast"] - t2_form["kast"]

        row[f"team1_player_kd_form_{last_n}"] = t1_form["kd"]
        row[f"team2_player_kd_form_{last_n}"] = t2_form["kd"]
        row[f"diff_player_kd_form_{last_n}"] = t1_form["kd"] - t2_form["kd"]

        row[f"team1_player_kddiff_form_{last_n}"] = t1_form["kddiff"]
        row[f"team2_player_kddiff_form_{last_n}"] = t2_form["kddiff"]
        row[f"diff_player_kddiff_form_{last_n}"] = t1_form["kddiff"] - t2_form["kddiff"]

        row[f"team1_player_wr_form_{last_n}"] = t1_form["wr"]
        row[f"team2_player_wr_form_{last_n}"] = t2_form["wr"]
        row[f"diff_player_wr_form_{last_n}"] = t1_form["wr"] - t2_form["wr"]

        row[f"team1_player_maps_form_{last_n}"] = t1_form["maps"]
        row[f"team2_player_maps_form_{last_n}"] = t2_form["maps"]
        row[f"diff_player_maps_form_{last_n}"] = t1_form["maps"] - t2_form["maps"]

        row[f"team1_player_form_score_{last_n}"] = t1_form["form_score"]
        row[f"team2_player_form_score_{last_n}"] = t2_form["form_score"]
        row[f"diff_player_form_score_{last_n}"] = (
            t1_form["form_score"] - t2_form["form_score"]
        )


    for last_n in [5, 10]:
        t1_map_form = get_team_player_form(
            team1_player_ids,
            game_time,
            last_n=last_n,
            map_name=map_name
        )

        t2_map_form = get_team_player_form(
            team2_player_ids,
            game_time,
            last_n=last_n,
            map_name=map_name
        )

        row[f"team1_player_map_form_score_{last_n}"] = t1_map_form["form_score"]
        row[f"team2_player_map_form_score_{last_n}"] = t2_map_form["form_score"]
        row[f"diff_player_map_form_score_{last_n}"] = (
            t1_map_form["form_score"] - t2_map_form["form_score"]
        )

        row[f"team1_player_map_wr_form_{last_n}"] = t1_map_form["wr"]
        row[f"team2_player_map_wr_form_{last_n}"] = t2_map_form["wr"]
        row[f"diff_player_map_wr_form_{last_n}"] = (
            t1_map_form["wr"] - t2_map_form["wr"]
        )

        row[f"team1_player_map_sample_{last_n}"] = t1_map_form["maps"]
        row[f"team2_player_map_sample_{last_n}"] = t2_map_form["maps"]
        row[f"diff_player_map_sample_{last_n}"] = (
            t1_map_form["maps"] - t2_map_form["maps"]
        )


    for m in [
        "Ancient",
        "Anubis",
        "Dust2",
        "Inferno",
        "Mirage",
        "Nuke",
        "Overpass",
        "Train",
        "Vertigo"
    ]:
        row[f"map_{m}"] = int(map_name == m)

    records.append(row)



model_df = pd.DataFrame(records)

model_df["team1_map_winrate"] = model_df["team1_map_winrate"].fillna(0.5)
model_df["team2_map_winrate"] = model_df["team2_map_winrate"].fillna(0.5)

model_df["diff_map_winrate"] = (
    model_df["team1_map_winrate"] -
    model_df["team2_map_winrate"]
)

numeric_cols = model_df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    if col != "team1_map_win":
        model_df[col] = model_df[col].replace([np.inf, -np.inf], np.nan)
        model_df[col] = model_df[col].fillna(model_df[col].median())

print("model_df shape:", model_df.shape)

print("\nTarget counts:")
print(model_df["team1_map_win"].value_counts())

print("\nTarget distribution:")
print(model_df["team1_map_win"].value_counts(normalize=True))

print("\nMissing values:")
print(model_df.isna().sum().sort_values(ascending=False).head(20))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    GradientBoostingClassifier,
    HistGradientBoostingClassifier
)
from sklearn.metrics import (
    roc_auc_score,
    log_loss,
    accuracy_score,
    roc_curve,
    confusion_matrix,
    ConfusionMatrixDisplay
)

try:
    import seaborn as sns
    sns.set_theme(style="whitegrid")
    USE_SEABORN = True
except ImportError:
    USE_SEABORN = False


model_df_sorted = model_df.sort_values("datetime").reset_index(drop=True)

y = model_df_sorted["team1_map_win"]

X = model_df_sorted.drop(
    columns=["team1_map_win", "match_id", "game_id", "datetime", "map_name"]
)

train_mask = (
    (model_df_sorted["datetime"] >= "2024-01-01") &
    (model_df_sorted["datetime"] < "2025-08-22")
)

test_mask = model_df_sorted["datetime"] >= "2025-08-22"

X_train = X.loc[train_mask].reset_index(drop=True)
X_test = X.loc[test_mask].reset_index(drop=True)

y_train = y.loc[train_mask].reset_index(drop=True)
y_test = y.loc[test_mask].reset_index(drop=True)

print("Train size:", len(X_train))
print("Test size:", len(X_test))
print("Train target distribution:")
print(y_train.value_counts(normalize=True))
print("Test target distribution:")
print(y_test.value_counts(normalize=True))


models = {
    "Decision Tree": DecisionTreeClassifier(
        random_state=42,
        max_depth=4,
        min_samples_leaf=80,
        min_samples_split=160
    ),

    "Random Forest": RandomForestClassifier(
        random_state=42,
        n_estimators=500,
        max_depth=5,
        min_samples_leaf=50,
        max_features="sqrt",
        n_jobs=-1
    ),

    "Extra Trees": ExtraTreesClassifier(
        random_state=42,
        n_estimators=500,
        max_depth=5,
        min_samples_leaf=50,
        max_features="sqrt",
        n_jobs=-1
    ),

    "Gradient Boosting": GradientBoostingClassifier(
        random_state=42,
        n_estimators=300,
        learning_rate=0.025,
        max_depth=3,
        min_samples_leaf=60,
        subsample=0.9
    ),

    "Hist Gradient Boosting": HistGradientBoostingClassifier(
        random_state=42,
        max_iter=400,
        learning_rate=0.025,
        max_leaf_nodes=15,
        max_depth=4,
        min_samples_leaf=60,
        l2_regularization=0.2
    ),

    "LightGBM": LGBMClassifier(
        random_state=42,
        objective="binary",
        verbose=-1,
        n_estimators=900,
        learning_rate=0.01,
        num_leaves=7,
        max_depth=3,
        min_child_samples=80,
        subsample=0.9,
        colsample_bytree=0.75,
        reg_alpha=0.1,
        reg_lambda=0.5
    )
}


results = []
predictions = {}

for name, model in models.items():
    print("Training:", name)

    model.fit(X_train, y_train)

    pred = model.predict_proba(X_test)[:, 1]
    pred_class = (pred >= 0.5).astype(int)

    predictions[name] = pred

    results.append({
        "model": name,
        "roc_auc": roc_auc_score(y_test, pred),
        "log_loss": log_loss(y_test, pred),
        "accuracy_0_5": accuracy_score(y_test, pred_class),
        "mean_pred": pred.mean(),
        "std_pred": pred.std(),
        "min_pred": pred.min(),
        "max_pred": pred.max()
    })

results_df = pd.DataFrame(results).sort_values("roc_auc", ascending=False)
display(results_df)


plt.figure(figsize=(10, 5))

plot_df = results_df.sort_values("roc_auc", ascending=True)

if USE_SEABORN:
    sns.barplot(data=plot_df, x="roc_auc", y="model")
else:
    plt.barh(plot_df["model"], plot_df["roc_auc"])

plt.xlabel("ROC AUC")
plt.ylabel("Model")
plt.title("Model comparison - ROC AUC")
plt.xlim(0.50, max(plot_df["roc_auc"]) + 0.03)

for i, row in enumerate(plot_df.itertuples()):
    plt.text(row.roc_auc + 0.002, i, f"{row.roc_auc:.4f}", va="center")

plt.grid(axis="x", alpha=0.3)
plt.show()


plt.figure(figsize=(10, 5))

plot_df = results_df.sort_values("log_loss", ascending=False)

if USE_SEABORN:
    sns.barplot(data=plot_df, x="log_loss", y="model")
else:
    plt.barh(plot_df["model"], plot_df["log_loss"])

plt.xlabel("Log loss")
plt.ylabel("Model")
plt.title("Model comparison - Log loss")

for i, row in enumerate(plot_df.itertuples()):
    plt.text(row.log_loss + 0.002, i, f"{row.log_loss:.4f}", va="center")

plt.grid(axis="x", alpha=0.3)
plt.show()


plt.figure(figsize=(10, 5))

plot_df = results_df.sort_values("accuracy_0_5", ascending=True)

if USE_SEABORN:
    sns.barplot(data=plot_df, x="accuracy_0_5", y="model")
else:
    plt.barh(plot_df["model"], plot_df["accuracy_0_5"])

plt.xlabel("Accuracy @ threshold 0.5")
plt.ylabel("Model")
plt.title("Model comparison - Accuracy @ 0.5")
plt.xlim(0.45, max(plot_df["accuracy_0_5"]) + 0.03)

for i, row in enumerate(plot_df.itertuples()):
    plt.text(row.accuracy_0_5 + 0.002, i, f"{row.accuracy_0_5:.4f}", va="center")

plt.grid(axis="x", alpha=0.3)
plt.show()


plt.figure(figsize=(8, 7))

for name, pred in predictions.items():
    fpr, tpr, _ = roc_curve(y_test, pred)
    auc = roc_auc_score(y_test, pred)
    plt.plot(fpr, tpr, label=f"{name} AUC = {auc:.4f}")

plt.plot([0, 1], [0, 1], linestyle="--", label="Random model")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC curves - tree models")
plt.legend()
plt.grid(alpha=0.3)
plt.show()


best_model_name = results_df.iloc[0]["model"]
best_pred = predictions[best_model_name]

plt.figure(figsize=(8, 5))

y_test_arr = np.asarray(y_test)

plt.hist(best_pred[y_test_arr == 0], bins=25, alpha=0.6, label="Team1 lost")
plt.hist(best_pred[y_test_arr == 1], bins=25, alpha=0.6, label="Team1 won")

plt.xlabel("Predicted probability of Team1 winning")
plt.ylabel("Number of maps")
plt.title(f"Prediction distribution - {best_model_name}")
plt.legend()
plt.grid(alpha=0.3)
plt.show()


best_class = (best_pred >= 0.5).astype(int)

cm = confusion_matrix(y_test, best_class)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Team1 lose", "Team1 win"]
)

disp.plot()
plt.title(f"Confusion matrix - {best_model_name}, threshold = 0.5")
plt.show()


for name in results_df["model"].head(3):
    model = models[name]

    if hasattr(model, "feature_importances_"):
        importance_df = pd.DataFrame({
            "feature": X_train.columns,
            "importance": model.feature_importances_
        }).sort_values("importance", ascending=False)

        top_features = importance_df.head(25).sort_values("importance")

        plt.figure(figsize=(10, 8))

        if USE_SEABORN:
            sns.barplot(data=top_features, x="importance", y="feature")
        else:
            plt.barh(top_features["feature"], top_features["importance"])

        plt.xlabel("Importance")
        plt.ylabel("Feature")
        plt.title(f"Top 25 feature importances - {name}")
        plt.grid(axis="x", alpha=0.3)
        plt.show()

        print("Top features -", name)
        display(importance_df.head(30))


print("Best model by ROC AUC:")
display(results_df.iloc[[0]])
```

## Conclusions

Adding recent player form helped the model.  
The ROC AUC improved from about **0.613** to about **0.630**.

Changing LightGBM parameters did not improve the result much.  
Most models stayed around **0.62–0.63 ROC AUC**, so the main limitation is probably the data/features, not the model settings.

Overall, the model has some predictive power, but it is not strong enough yet to consistently reach 0.70+ ROC AUC. 
To improve it further, the best next step would be adding stronger team-level features, such as team Elo, recent team form, map win rate by team, tournament tier, or pre-match rankings.

## Decision Tree Models Comparison

Several tree-based models were tested, including Decision Tree, Random Forest, Extra Trees, Gradient Boosting, Hist Gradient Boosting, and LightGBM.

Overall, LightGBM and simpler boosted models performed the best.  
Single Decision Tree was weaker, because it is too simple and can easily miss more complex patterns.

Random Forest and Extra Trees were useful for comparison, but they did not clearly outperform LightGBM.

The results show that boosted tree models are the best choice for this dataset, but the difference between models was not huge. This means that further improvement will probably come more from better features than from changing the model type.
